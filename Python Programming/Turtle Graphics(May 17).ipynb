### Problem 2: 


#### Problem Statement
    Plot a line graph on income vs year for the state of California
#### Constraints

#### Test Cases

```python
import pandas as pd
import numpy as np

#t = np.arange(2005,2013)
df = pd.read_csv('DataFiles/Income.csv')
y = df.values[4]
x = df.columns
years = []
income = []
for i in range(2, len(x)):
    years.append(x[i])
for i in range(2,len(y)):
    income.append(y[i])
plt.xlabel('Year')
plt.grid(True)
plt.ylabel('income')
plt.plot(years,income)
plt.title('Income of California State')
plt.savefig('Figures/income.png')
plt.show()
```

### Problem 3: 


#### Problem Statement
    Plot a bar graph on income vs year for the state of California
#### Constraints

#### Test Cases

```python
width = 1/2
plt.bar(years,income,width,color=('red','blue','yellow','green'))
plt.title('Income of California State')
plt.savefig('Figures/income-bar.png')
```

### Problem 4: 


#### Problem Statement
    Generate a scatter plot for 50 random numbers.
#### Constraints

#### Test Cases

```python
n = 50
x = np.random.rand(n) #Generate 50 random numbers
y = np.random.rand(n) #Generate 50 random numbers
colorlist = np.random.rand(n) #Generate 50 random color values
arealist = np.pi * (15*np.random.rand(50))**2 #Generate 50 random areas of circles
plt.scatter(x,y,s = arealist,c = colorlist,alpha =2 )
plt.savefig('Figures/scatter-plot-50-random-numbers.png')
```
